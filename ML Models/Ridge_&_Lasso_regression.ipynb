#Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.State=le.fit_transform(df.State)
```

```python
df.head()
```

#X & Y split

```python
#x and y split
x=df.drop(columns=['Profit'],axis=1)
x.head()
```

```python
y=df.Profit
y.head()
```

#Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
```

```python
x_scaled=pd.DataFrame(scale.fit_transform(x),columns=x.columns)
x_scaled.head()
```

#Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

#Model Building

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
r=Ridge()
l=Lasso()
```

```python
r.fit(x_train,y_train)
```

```python
l.fit(x_train,y_train)
```

```python
pred1= r.predict(x_test)
pred1
```

```python
pred1_train = r.predict(x_train)
```

```python
pred2= l.predict(x_test)
pred2
```

```python
pred2_train = l.predict(x_train)
pred2_train
```

```python
profit=pd.DataFrame({'Actual Profit':y_test,'Ridge_pred':pred1,'Lasso_pred':pred2})
profit.head()
```

#Evaluate Ridge and Lasso Models

```python
from sklearn import metrics
```

```python
#r2 scpre
print(metrics.r2_score(y_test,pred1))
print(metrics.r2_score(y_test,pred2))
```

```python
#MSE
print(metrics.mean_squared_error(y_test,pred1))
print(metrics.mean_squared_error(y_test,pred2))
```

```python
#RMSE
print(np.sqrt(metrics.mean_squared_error(y_test,pred1)))
print(np.sqrt(metrics.mean_squared_error(y_test,pred2)))
```
