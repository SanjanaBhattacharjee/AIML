#Outlier Detection

```python
df.head()
```

```python
sns.boxplot(df.bmi)
```

```python
#removal methods- IQR(Inter Quatile Range), Z-score, Percentile method
```

```python
#Outlier removal by replacement with median
```

```python
q1=df.bmi.quantile(0.25) #Q1
q3=df.bmi.quantile(0.75) #Q3
```

```python
print(q1)
print(q3)
```

```python
IQR=q3-q1
```

```python
IQR
```

```python
upper_limit=q3+1.5*IQR
```

```python
upper_limit
```

```python
lower_limit=q3-1.5*IQR
```

```python
lower_limit
```

```python
df.median()
```

```python
df['bmi']= np.where(df['bmi']>upper_limit,30,df['bmi'])
```

```python
sns.boxplot(df.bmi)
```

#1)Outlier removal using IQR

```python
q1=df.bmi.quantile(0.25) #Q1
q3=df.bmi.quantile(0.75) #Q3
```

```python
IQR=q3-q1
```

```python
upper_limit= q3+1.5*IQR
```

```python
upper_limit
```

```python
df=df[df.bmi<upper_limit]
```

```python
sns.boxplot(df.bmi)
```

```python
df.shape
```

#2)Outlier removal using Z-Score

```python
from scipy import stats
```

```python
sns.boxplot(df.bmi)
```

```python
bmi_zscore=stats.zscore(df.bmi)
bmi_zscore
```

```python
df_z=df[np.abs(bmi_zscore)<=3]
```

```python
sns.boxplot(df_z.bmi)
```

```python
df_z.shape
```

# 3)Outlier removal using Percentile

```python
sns.boxplot(df.bmi)
```

```python
p99=df.bmi.quantile(0.99)
```

```python
p99
```

```python
df= df[df.bmi<=p99]
```

```python
sns.boxplot(df.bmi)
```
