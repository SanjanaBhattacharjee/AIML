```python
# import required libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Breast Cancer Wisconsin (Diagnostic) Data Set.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['id','Unnamed: 32'],axis = 1,inplace =True)
```

```python
df.head()
```

```python
df.diagnosis.value_counts()
```

```python
X= df.drop(['diagnosis'],axis =1)
```

```python
y = df.diagnosis
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
```

```python
y =le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
# Creating ANN structure

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
classification = Sequential()
classification.add(Dense(30,activation='relu'))
classification.add(Dense(128,activation='relu'))
classification.add(Dense(64,activation='relu'))
classification.add(Dense(32,activation='relu'))
classification.add(Dense(1,activation='sigmoid'))
```

```python
# Compiling ANN model

classification.compile(optimizer='adam',loss ='binary_crossentropy',metrics=['accuracy'] )
```

```python
# training the model
classification.fit(x_train,y_train,batch_size=10,epochs =300)
```

```python
# testing the model

y_pred = classification.predict(x_test)
```

```python
y_pred
```

```python
y_pred =y_pred.flatten().astype('int')
```

```python
# Comparing the actual and predicted y values

pd.DataFrame({'Actual y_value':y_test.flatten(),'Predicted y_value':y_pred.flatten().astype('int')}).head(10)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
